```python
# CSV 파일 경로
file_path = 'D:\\MJ 데이터 공모전 로우 데이터\\origins\\20_origin_df.csv'

# CSV 파일 불러오기
origin_df = pd.read_csv(file_path, encoding='utf-8', header=0)
```

```python
copied_df = origin_df.copy()
```

```python
copied_df = copied_df['0'].str.replace('국정조사_', '').str.replace('국정감사_', '').str.replace('개회식_', '')

copied_df = pd.DataFrame(copied_df)

copied_df = copied_df['0'].str.split('회_', expand=True)

copied_df = copied_df.drop([0, 1], axis=1)

copied_df = copied_df[2].str.split('.csv 0', expand=True)

copied_df[0] = copied_df[0].apply(lambda x: re.sub(r'\d{1,2}차_', '', x))

copied_df = pd.DataFrame(copied_df)

# 정규식 패턴과 숫자를 삭제하는 함수 정의
def remove_numbers_in_parentheses(text):
    import re
    pattern = r"\(\d+\)"
    return re.sub(pattern, '', str(text))

copied_df[0] = copied_df[0].apply(remove_numbers_in_parentheses)

copied_df = pd.DataFrame(copied_df)

copied_df[0] = copied_df[0].str.rstrip()

copied_df = copied_df.sort_values(by=copied_df.columns[0])
```

```python
def extract_date(s):
    first_word = s.split()[0]  # 첫번째 단어 추출
    match = re.search(r'\d{4}년\d{1,2}월\d{1,2}일', first_word)  # 일자 형식의 문자열 찾기
    return match.group() if match else None  # 일자가 있으면 반환, 없으면 None 반환

copied_df['date'] = copied_df[1].apply(extract_date)
```

```python
# 16대의 경우 '최용규의원'처럼 규칙에 어긋난 리스트의 원소들이 있어서 이를 '최용규 의원'으로 바꾸는 코드를 만들었습니다.

def process_text(text, names):
    first_word = text.split()[0] # 첫 번째 단어를 찾는다
    for name in names:
        if name in first_word and '위원' in first_word: 
            # 첫 번째 단어에 'names'의 원소와 '위원'이 포함되어 있는지 확인한다
            new_word = first_word.replace(name, f"{name} ")
            # 첫 번째 단어에 원소가 포함되어 있고 '위원'이라는 문자열이 포함되어 있는 경우 사이에 공백을 추가한다
            text = text.replace(first_word, new_word)
    return text

grouped_df.iloc[:, 1] = grouped_df.iloc[:, 1].apply(lambda x: [process_text(text, names) for text in x])
```

```python
import pandas as pd
from collections import defaultdict

# 단어와 내용을 분리하는 함수 정의
def split_speaker_content(content_list):
    speakers = []
    contents = []
    for content in content_list:
        split_content = content.split(' ', 2) # 첫 두 단어는 '이름' '위원'이므로 세 번째 단어부터가 내용
        speakers.append(split_content[0])
        contents.append(split_content[2] if len(split_content) > 2 else '')
    return speakers, contents

# 위원회, 발언 내용을 수집하는 defaultdict 생성
speaker_info = defaultdict(lambda: {'committees': set(), 'contents': []})

# 각 로우에 대해 처리
for idx, row in merged_final_df.iterrows():
    speakers, contents = split_speaker_content(row['Content'])
    for speaker, content in zip(speakers, contents):
        speaker_info[speaker]['committees'].add(row['Committee'])
        speaker_info[speaker]['contents'].append(content)

# 결과 데이터프레임 생성
result_df = pd.DataFrame({
    'Name': list(speaker_info.keys()),
    'Committees': [list(info['committees']) for info in speaker_info.values()],
    'Contents': [info['contents'] for info in speaker_info.values()],
})
```

```python
result_df['Content_Length'] = result_df['Contents'].apply(len)
result_df = result_df.sort_values('Content_Length', ascending=True)
result_df.reset_index(drop=True, inplace=True)
result_df
```
