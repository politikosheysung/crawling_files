```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 크롬드라이버 서비스를 설정합니다.
service = Service('./chromedriver')
driver = webdriver.Chrome(service=service)

# 저장할 디렉터리 이름
output_dir = "TMP" # 더불어민주당(The Minjoo party)

if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
# 데이터를 저장할 DataFrame 생성
end_of_category_df = pd.DataFrame(columns=['content'])
    
# 페이지 접속
driver.get('https://theminjoo.kr/main/sub/news/view.php?brd=11&post=103338')

end_of_category_content_xpath = "/html/body/main/div/div[2]/div/div/div[1]/div[2]/div[1]"
end_of_category_content_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, end_of_category_content_xpath)))
content = end_of_category_content_element.text.strip()
end_row = {'content': content}
end_of_category_df = pd.concat([end_of_category_df, pd.DataFrame([end_row])], ignore_index=True)

end_of_briefing = end_of_category_df.iloc[0,0]
```

```python
# 크롬드라이버 서비스를 설정합니다.
service = Service('./chromedriver')
driver = webdriver.Chrome(service=service)

# 저장할 디렉터리 이름
output_dir = "TMP" # 더불어민주당(The Minjoo party)

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 페이지 접속
driver.get('https://theminjoo.kr/main/sub/news/list.php?sno=0&par=&&par=&brd=11')
    
# 데이터를 저장할 DataFrame 생성
TMP_scraped_df = pd.DataFrame(columns=['date', 'content'])

while True:
    if content == end_of_briefing:
        driver.quit()
        break
    for brifing_num in range(1, 21):
        # 카테고리에 있는 브리핑을 모두 클릭
        briefing_xpath = f'/html/body/main/div/div[3]/div/div[2]/div[1]/div[{brifing_num}]/a'
        briefing_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, briefing_xpath)))
        briefing_element.click()

        # 작성일자와 대변인 발언 내용 크롤링
        date_xpath = "/html/body/main/div/div[2]/div/div/div[1]/div[1]/ul/li[3]"
        content_xpath = "/html/body/main/div/div[2]/div/div/div[1]/div[2]"
        date_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, date_xpath)))
        content_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, content_xpath)))
        date = date_element.text.strip()
        content = content_element.text.strip()

        # 정보를 DataFrame에 추가
        new_row = {'date': date, 'content': content}
        TMP_scraped_df = pd.concat([TMP_scraped_df, pd.DataFrame([new_row])], ignore_index=True)

        to_back_xpath = "/html/body/main/div/div[2]/div/div/div[2]/a"
        back_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, to_back_xpath)))
        back_element.click()

        # 명시적 대기를 통해 briefing_element의 요소가 화면에 나타날 때까지 대기(최대 10초)
        briefing_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, briefing_xpath)))

        if brifing_num == 20:
            next_page_xpath = "/html/body/main/div/div[3]/div/div[2]/div[2]/a[12]"
            next_page_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, next_page_xpath)))
            next_page_element.click()

TMP_scraped_df.to_csv(os.path.join(output_dir, 'TMP_scraped_data.csv'), index=False, encoding='utf-8')
```
